```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Set 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
# Create a bar chart without annotations. 
x_axis = np.arange(0, len(gyms))
plt.bar(x_axis, members)

plt.title("Popular Gyms")
plt.xlabel("Gym")
plt.ylabel(" # of Members")
plt.xticks(x_axis, gyms)
```
